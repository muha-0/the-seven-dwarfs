Libraries

```python
import pandas as pd
import numpy as np
import cv2
import os
```

```python
Getting the directories where data exist
```

```python
training_data_dir_cancer = r'E:\my ai projects\binary_classification\Skin_Data\Cancer\Training'
training_data_dir_no_cancer = r'E:\my ai projects\binary_classification\Skin_Data\Non_Cancer\Training'
test_data_dir_cancer = r'E:\my ai projects\binary_classification\Skin_Data\Cancer\Testing'
test_data_dir_no_cancer = r'E:\my ai projects\binary_classification\Skin_Data\Non_Cancer\Testing'
X_train = []
X_test = []
y_train = []
y_test = []
```

Reading the data

```python
# Cancer training

for file in os.listdir(training_data_dir_cancer):
    img = cv2.imread(os.path.join(training_data_dir_cancer,file))
    if img is not None:
        img_resized = cv2.resize(img, (200,200))
        X_train.append(img_resized.flatten())
        y_train.append(1)
    else:
        print("oops")

# Cancer test

for file in os.listdir(test_data_dir_cancer):
    img = cv2.imread(os.path.join(test_data_dir_cancer,file))
    if img is not None:
        img_resized = cv2.resize(img, (200,200))
        X_test.append(img_resized.flatten())
        y_test.append(1)
    else:
        print("oops")

# Non Cancer training

for file in os.listdir(training_data_dir_no_cancer):
    img = cv2.imread(os.path.join(training_data_dir_no_cancer,file))
    if img is not None:
        img_resized = cv2.resize(img, (200,200))
        X_train.append(img_resized.flatten())
        y_train.append(0)
    else:
        print("oops")


# Non Cancer test

for file in os.listdir(test_data_dir_no_cancer):
    img = cv2.imread(os.path.join(test_data_dir_no_cancer,file))
    if img is not None:
        img_resized = cv2.resize(img, (200,200))
        X_test.append(img_resized.flatten())
        y_test.append(0)
    else:
        print("oops")
```

Shuffling and normalizing

```python
from sklearn.utils import shuffle

X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)

X_train, y_train = shuffle(X_train, y_train, random_state = 42)

X_train = X_train/255.0
X_test = X_test/255.0
```

Training :)))

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)
```

Testing the accuracy. 76% is good for that amount of data so cool

```python
predicted = model.predict(X_test)

print("Accuracy: ",end = '')

print(((y_test == predicted).sum())/len(y_test))
```
